- En este caso (dataset astronomico) es complicado determinar valores atipicos, ya que sus valores corresponden a su ubicacion, brillo y tamao en el espacio, por lo cual, estos valores representan interes de estudio astronomico.
- Se analizaran los quantiles Q1 y Q3 para tener una mejor idea del caso

```python
# Funcion de calculo quantiles, Q1, Q3 y valores limites para cada clase
def quantilQ1Q3(data): 
    Q1 = data.quantile(q=0.25)
    Q3 = data.quantile(q=0.75)
    IQR=Q3-Q1
    minlimit= Q1 - 1.5 * IQR
    maxlimit= Q3 + 1.5 * IQR 
    return minlimit, maxlimit
```

```python
# Calculo quantiles, Q1, Q3 y valores limites para cada clase. Creacion de dataFrame valores limites
def getLimit(astro):
    cols = ['u', 'g', 'r', 'i', 'z', 'ra', 'dec', 'redshift']
    temp1=pd.DataFrame(['min', 'max'], columns=["minmax"])
    data2 =df[df['class']==astro][cols].copy()
    ast=pd.DataFrame([astro, astro], columns=["class"])
    qlimit=pd.concat([ast,temp1], axis=1)
    for var in cols:
        rows=quantilQ1Q3(data2[var])
        temp2=pd.DataFrame(rows, columns=[var])
        qlimit=pd.concat([qlimit, temp2], axis=1)
    return qlimit

qlimitStar=getLimit('STAR')
qlimitQSO=getLimit('QSO')
qlimitGalaxy=getLimit('GALAXY')
qlimit=pd.concat([qlimitStar, qlimitQSO, qlimitGalaxy], keys=["STAR", "QSO", "Galaxy"])

qlimit
```

- Considerando que utilizar el metodo de quantiles para determinar outliers se ocupa sobre distribuciones normales, realizaremos un test de normalidad

### Test de Normalidad ###

```python
# test de normalidad 'u', 'g', 'r', 'i', 'z', 'ra', 'dec', 'redshift'
cols = ['u', 'g', 'r', 'i', 'z', 'ra', 'dec', 'redshift']
for astro in df['class'].unique().tolist():
    data2 =df[df['class']==astro][cols].copy()
    for var in cols:
        stat, p = normaltest(data2[var])
        print(astro, 'var=', var)
        # interpretacion
        alpha = 1e-3
        if p < alpha:
            print('Statistics = %.3f, p = %g,  does NOT look like a normal distribution, H0 can be rejected' % (stat, p))
        else: 
            print('Statistics = %.3f, p = %g,  looks like a normal distribution, H0 cannot be rejected' % (stat, p))
```

- Ninguna de la variables muestra una distribucion normal segun el test aplicado, por lo tanto no podemos eliminar outliers con el metodo de quantiles

### Normalizacion de variables

```python
# Histograma redshift clases QSO y Galaxy 
plt.figure(figsize=(8, 6))
df2=df[df['class']!='STAR'].copy()
sns.histplot(data=df2, x=zscore(df2['redshift']), legend=True, hue='class', multiple='stack');
plt.xlim(-1, 1)
plt.ylim(0, 2500)
plt.title('Histograma de Densidad redshift', size=14);
```

- No se utiliz la variable STAR, ya que se observ en el grafico anterio que sus valores estan en torno a cero.
- Como ya se habia dicho, esta distribucion no corresponde a una normal, por tanto la funcion de normalizacion zscore, no es aplicable. Se observa que los valores centrales se encuentran en torno a -0.35.
- Se aplicara mas adelante la funcion tanh para obtener una distribucion normal de estos valores 

```python
# Grafico de frecuencias de STAR, QSO y Galaxia segun las variables espectrales u, g, r, i, z utilizando zscore
cols = ['u', 'g', 'r', 'i', 'z']
plt.figure(figsize=(20,6))
sns.set_style("ticks")
j=1
for var in cols:
    plt.subplot(1, 5, j)
    sns.histplot(data=df, x=zscore(df[var]), legend=True, hue='class'); 
    j+=1
    plt.xlabel(var, size=15)   
plt.tight_layout()
```

```python
# Grafico de frecuencias de u, g, r, i, z, segun tipo de astro utilizando zscore
cols = ['u', 'g', 'r', 'i', 'z']
plt.figure(figsize=(20,6));
j=1
for astro in df['class'].unique().tolist():
    plt.subplot(1, 3, j)
    i=0
    for col in cols:
        data2=df[df['class']==astro]
        sns.histplot(data=data2,x=zscore(data2[col]), label =col, legend=True, kde=False, color=sns.color_palette()[i],
                    fill=False, element="poly", stat="density")
        plt.xlabel(astro, size=15)
        i+=1
    j+=1
    plt.legend()
    plt.tight_layout();
```

- Se observa de los graficos que la medida central esta en el valor cero para todas las variables (normalizando as los datos)


### Normalizacion variable redshift ###

```python
# Histograma redshift clases QSO y Galaxy 
pg=0.15
pq=2.5
ps=0.0017
plt.figure(figsize=(15, 5))
plt.suptitle('Normalizacion variable redshift', size=14);
plt.subplot(1, 3, 1)

df2=df[(df['class']=='GALAXY')].copy()
sns.histplot(data=df2, x=np.tanh((df2['redshift'])/pg), legend=True, hue='class', multiple='stack');
plt.subplot(1, 3, 2)
df2=df[(df['class']=='QSO')].copy()
sns.histplot(data=df2, x=np.tanh((df2['redshift'])/pq), legend=True, hue='class', multiple='stack');
plt.subplot(1, 3, 3)
df2=df[(df['class']=='STAR')].copy()
sns.histplot(data=df2, x=np.tanh((df2['redshift'])/ps), legend=True, hue='class', multiple='stack');
plt.tight_layout();
```

### Normalizacion variable redshift y standarizacion zscore ###

```python
# Histograma redshift clases QSO y Galaxy 
pg=0.15
pq=2.5
ps=0.0017
plt.figure(figsize=(16, 5))
plt.suptitle('Normalizacion variable redshift y standarizacion zscore', size=18);
plt.subplot(1, 3, 1)

df2=df[(df['class']=='GALAXY')].copy()
sns.histplot(data=df2, x=zscore(np.tanh((df2['redshift'])/pg)), legend=True, hue='class', multiple='stack');
plt.subplot(1, 3, 2)
df2=df[(df['class']=='QSO')].copy()
sns.histplot(data=df2, x=zscore(np.tanh((df2['redshift'])/pq)), legend=True, hue='class', multiple='stack');
plt.subplot(1, 3, 3)
df2=df[(df['class']=='STAR')].copy()
sns.histplot(data=df2, x=zscore(np.tanh((df2['redshift'])/ps)), legend=True, hue='class', multiple='stack');
plt.tight_layout();
```

- Se puede observar que luego de los procesos normalizacion y standarizacion zscore las distribuciones (excepto la clase QSO) se asemejan a normales y centradas en cero

### Matriz de corelacion para cada clase ###

```python
#cols = ['u', 'g', 'r', 'i', 'z']
sns.set(font_scale=1.1)
fig, axs = plt.subplots(3, 1, sharex = False, figsize=(20,15))
i=0
for astro in df['class'].unique().tolist():
    axs[i].set_title('Correlacion {}'.format(astro), size=14)
    data2=df[df['class']==astro].drop('rerun', axis=1, inplace=False)
    corr = data2.corr()
    mask = np.tril(np.ones_like(corr, dtype=np.bool))
    sns.heatmap(corr, annot=True, cmap='BrBG', mask=mask, ax=axs[i], vmin=-1, vmax=1);
    #sns.heatmap(corr, annot=True, cmap='BrBG', mask=None, ax=axs[i], vmin=-1, vmax=1);
    i+=1
    plt.tight_layout(pad=3);
    
```

- En general las correlaciones son similares, se observan alguna diferencias en los parametros espectrales (u,g,r,i,z), sin embargo a simple vista no son significativas

### Matriz de corelacion utilizando las clases ###

- Como la clase es un valor categorico, lo convertiremos a numerico usando la funcion one-hot encoder

```python
# one-hot in class
#df_hot=pd.get_dummies(df, columns=['class'], drop_first=False).drop('rerun', axis=1, inplace=False) 
df_hot = pd.concat([df.drop('rerun', axis=1, inplace=False), pd.get_dummies(df['class'],drop_first=False)],axis=1)
df_hot.head()
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=1)
corr = df_hot.corr()
plt.title('Correlacion GALAXY', size=14);
mask = np.tril(np.ones_like(corr, dtype=np.bool))
sns.heatmap(corr, annot=True, cmap='BrBG', vmin=-1, vmax=1, mask=mask);
```

- Se que los valores de luminosidad (u,g,r,i,z) estan fuertemente correlacionados (como era de esperar)
- Existe fuerte correlacion entre galaxia y estrella, no asi con quasar
- Los quasars estan fuertemente correlacionados con la variable redshift
